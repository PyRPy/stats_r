```r
# before you use for loop, think about 'lappy()' 
# especially when the length of vectors are not same
```

```r
myList = list(x = c(1, 2, 3), 
              y = c(1, 2, 3, 4), 
              z = c(1, 2, 3, 4, 5))
```

```r
list_sum = lapply(myList, sum)
```

```r
list_sum$z
```

```r
list_mean = lapply(myList, mean)
```

```r
list_mean
```

```r
# or you can extract mean of y
list_mean$y
```
