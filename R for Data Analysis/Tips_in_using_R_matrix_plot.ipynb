```r
# how to quickly plot columns in a matrix
```

```r
dat = matrix(rnorm(24) + 1, nrow = 12)
```

```r
dat
```

```r
colnames(dat) = c("first", "second")
```

```r
rownames(dat) = month.abb
```

```r
dat
```

```r
matplot(dat, type = "l", 
        main = "Quick matrix plot")
legend("topright", legend = colnames(dat), 
        pch = 5:6, col = 1:2, cex=0.8)
axis(1, at = 1:12, labels = rownames(dat))
```
