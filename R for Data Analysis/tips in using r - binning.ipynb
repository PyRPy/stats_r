```r
# how to 'bin' the numbers
```

```r
# create some data

my_vector0 = c()
my_vector1 = c()
for (i in 1:5) { 
  my_vector1 = sample(1:15, 5)
  my_vector0 = c(my_vector0, my_vector1)  
}
```

```r
# before binning
table(my_vector0)
```

```r
# let's use dplyr
library(dplyr)
```

```r
my_table = data.frame(numbers = my_vector0)
```

```r
my_table[1:5,]
```

```r
freqency_table <- my_table %>% 
  mutate(new_bin = cut(numbers, breaks = c(0, 5, 10, 15)))
```

```r
table(freqency_table$new_bin)
```
