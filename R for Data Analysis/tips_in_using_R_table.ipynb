```r
# how to extract data from a table()
```

```r
# create some random numbers
```

```r
my_vector0 = c()
my_vector1 = c()
```

```r
for (i in 1:5) { 
  my_vector1 = sample(1:10, 5)
  my_vector0 = c(my_vector0, my_vector1)  
}
```

```r
table(my_vector0)
```

```r
# extract frequencies
as.numeric(table(my_vector0))
```

```r
# extract names
names(table(my_vector0))
```
